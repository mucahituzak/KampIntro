**Veri Temizleme**

---
**Eksik Değerlerle Başa Çıkma**


Bazı makine öğrenimi modelleri eksik değerlere sahip verileri kabul etmediği için eksik değerler sorunuyla ilgilenmeliyiz. İlk olarak, veri kümemizdeki eksik değerlerin sayısını görelim. Gerçekte eksik değerler içeren her sütun için eksik değerlerin sayısını ve yüzdesini görmek istiyoruz.

```python
# Her sütundaki eksik değerlerin sayısını alma
num_missing = dataset.isna().sum()
# Eksik 0 değer içeren sütunların hariç tutulması
num_missing = num_missing[num_missing > 0]
# Kayıp değerlerin yüzdelerini alma
percent_missing = num_missing * 100 / dataset.shape[0]
# Kayıp değerlerin sayısı ve yüzdesinin birleştirilmesi
# tek bir veri çerçevesi haline getirme ve sıralama
pd.concat([num_missing, percent_missing], axis=1,
          keys=['Missing Values', 'Percentage']).\
          sort_values(by="Missing Values", ascending=False)
```

Şimdi bu kayıp değerlerle uğraşmaya başlıyoruz.
