**Eksik Değerlerle Başa Çıkma**


Bazı makine öğrenimi modelleri eksik değerlere sahip verileri kabul etmediği için eksik değerler sorunuyla ilgilenmeliyiz. İlk olarak, veri kümemizdeki eksik değerlerin sayısını görelim. Gerçekte eksik değerler içeren her sütun için eksik değerlerin sayısını ve yüzdesini görmek istiyoruz.

```python
# Her sütundaki eksik değerlerin sayısını alma
num_missing = dataset.isna().sum()
# Eksik 0 değer içeren sütunların hariç tutulması
num_missing = num_missing[num_missing > 0]
# Kayıp değerlerin yüzdelerini alma
percent_missing = num_missing * 100 / dataset.shape[0]
# Kayıp değerlerin sayısı ve yüzdesinin birleştirilmesi
# tek bir veri çerçevesi haline getirme ve sıralama
pd.concat([num_missing, percent_missing], axis=1,
          keys=['Missing Values', 'Percentage']).\
          sort_values(by="Missing Values", ascending=False)
```

Şimdi bu kayıp değerlerle uğraşmaya başlıyoruz.

**Pool QC**

---

Havuz  sütunundaki eksik değerlerin yüzdesi %99,56'dır ve bu çok yüksek bir orandır. Bu sütundaki eksik değerin ilgili evin havuzu olmadığını gösterdiğini düşünüyoruz. Bunu doğrulamak için Havuz Alanı sütunundaki değerlere bir göz atalım:

```python
dataset["Pool Area"].value_counts()
```

Havuz Alanı sütununda 0 değerine sahip 2917 girdi olduğunu görebiliriz. Bu, havuzu olmayan her evin Havuz QC sütununda eksik bir değere ve Havuz Alanı sütununda 0 değerine sahip olduğu hipotezimizi doğrulamaktadır. O halde Havuz QC sütunundaki eksik değerleri "Havuz Yok" ile dolduralım:

```python
dataset["Pool QC"].fillna("No Pool", inplace=True)
```

**Misc Feature**

---

Misc Val sütunundaki eksik değerlerin yüzdesi %96,38'dir ve bu da çok yüksektir.  Misc Val sütunundaki değerlere bir göz atalım:


```python
dataset["Misc Val"].value_counts()
```

Misc Val sütununda 0 değerine sahip 2827 girdi olduğunu görebiliriz. Misc Feature sütununda ise 2824 eksik değer vardır. O halde, Pool QC'de olduğu gibi, "çeşitli özelliği" olmayan her evin Misc Feature sütununda eksik bir değere ve Misc Val sütununda 0 değerine sahip olduğunu söyleyebiliriz. O halde Misc Feature sütunundaki eksik değerleri "No Feature" ile dolduralım:

```python
dataset['Misc Feature'].fillna('No feature', inplace=True)
```

**Alley, Fence, ve Fireplace Qu**

---

Veri seti belgelerine göre, Alley, Fence ve Fireplace Qu sütunlarındaki NA, evin bir sokağı, çiti veya şöminesi olmadığını gösterir. Dolayısıyla, bu sütunlardaki eksik değerleri uygun şekilde "Sokak Yok", "Çit Yok" ve "Şömine Yok" ile dolduruyoruz:


```python
dataset['Alley'].fillna('No Alley', inplace=True)
dataset['Fence'].fillna('No Fence', inplace=True)
dataset['Fireplace Qu'].fillna('No Fireplace', inplace=True)
```

**Lot Frontage**

---
Daha önce gördüğümüz gibi, Arsa Cephesi eve bağlanan sokağın doğrusal ayaklarını temsil eder. Dolayısıyla, bu sütundaki eksik değerlerin evin herhangi bir sokağa bağlı olmadığını gösterdiğini varsayıyoruz ve eksik değerleri 0 ile dolduruyoruz:


```python
dataset['Lot Frontage'].fillna(0, inplace=True)
```

**Garage Cond, Garage Qual, Garage Finish, Garage Yr Blt, Garage Type, Garage Cars, ve Garage Area**

---
Veri kümesi belgelerine göre, Garage Cond, Garage Qual, Garage Finish ve Garage Type sütunlarında NA olması evde garaj olmadığını gösterir. Bu yüzden bu sütunlardaki eksik değerleri "Garaj Yok" ile dolduruyoruz. Garage Cond, Garage Qual, Garage Finish, Garage Yr Blt sütunlarında 159 eksik değer olduğunu, ancak Garage Type'da 157 ve hem Garage Cars hem de Garage Area'da bir eksik değer olduğunu fark ediyoruz. Garage Cars'taki eksik değeri içeren satıra bir göz atalım:


```python
garage_columns = [col for col in dataset.columns if col.startswith("Garage")]
dataset[dataset['Garage Cars'].isna()][garage_columns]
```

Bunun Garage Area' daki eksik değeri içeren satırla aynı olduğunu ve Garage Type dışındaki tüm garaj sütunlarının bu satırda boş olduğunu görebiliriz, bu nedenle Garage Cars ve Garaj Alanı'ndaki eksik değerleri 0 ile dolduracağız.

<br/>

Garage Cond, Garage Qual, Garage Finish ve Garage Yr Blt sütunları null iken Garage Type'ın null olmadığı 2 satır olduğunu gördük. Bu iki satıra bir göz atalım:

```python
dataset[~pd.isna(dataset['Garage Type']) &
        pd.isna(dataset['Garage Qual'])][garage_columns]
```

Bu iki satırdaki Garaj Tipi değerlerini de "Garaj Yok" ile değiştireceğiz.

Garage Yr Blt için, bu sayısal bir sütun olduğu için eksik değerleri 0 ile dolduracağız:

```python
dataset['Garage Cars'].fillna(0, inplace=True)
dataset['Garage Area'].fillna(0, inplace=True)

dataset.loc[~pd.isna(dataset['Garage Type']) &
            pd.isna(dataset['Garage Qual']), "Garage Type"] = "No Garage"

for col in ['Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond']:
    dataset[col].fillna('No Garage', inplace=True)

dataset['Garage Yr Blt'].fillna(0, inplace=True)
```

**Bsmt Exposure, BsmtFin Type 2, BsmtFin Type 1, Bsmt Qual, Bsmt Cond, Bsmt Half Bath, Bsmt Full Bath, Total Bsmt SF, Bsmt Unf SF, BsmtFin SF 2, ve BsmtFin SF 1**

---
Veri seti dokümantasyonuna göre, bu sütunların ilk beşinden herhangi birinde NA olması evde bodrum olmadığını gösterir. Bu yüzden bu sütunlardaki eksik değerleri "Bodrum Yok" ile dolduruyoruz. Bu sütunların ilk beşinde 80 eksik değer olduğunu, ancak BsmtFin Type 2'de 81, Bsmt Exposure'da 83, Bsmt Half Bath ve Bsmt Full Bath'ın her birinde 2 ve diğerlerinin her birinde 1 olduğunu fark ediyoruz:


```python
bsmt_columns = [col for col in dataset.columns if "Bsmt" in col]
dataset[dataset['Bsmt Half Bath'].isna()][bsmt_columns]
```

Bunların Bsmt Full Bath'daki eksik değerleri içeren satırlarla aynı olduğunu ve bu iki satırdan birinin Total Bsmt SF, Bsmt Unf SF, BsmtFin SF 2 ve BsmtFin SF 1 sütunlarının her birinde eksik değer içerdiğini görebiliriz. Ayrıca, Bsmt Exposure, BsmtFin Type 2, BsmtFin Type 1, Bsmt Qual ve Bsmt Cond'un bu satırlarda boş olduğunu fark ediyoruz, bu nedenle Bsmt Half Bath, Bsmt Full Bath, Total Bsmt SF, Bsmt Unf SF, BsmtFin SF 2 ve BsmtFin SF 1 sütunlarındaki eksik değerleri 0 ile dolduracağız.

<br/>

Bsmt Exposure'ın null olduğu ancak BsmtFin Type 1, Bsmt Qual ve Bsmt Cond'ın null olmadığı 3 satır olduğunu gördük. Şimdi bu üç satıra bir göz atalım:

```python
dataset[~pd.isna(dataset['Bsmt Cond']) &
        pd.isna(dataset['Bsmt Exposure'])][bsmt_columns]
```

Bu üç satır için Bsmt Exposure'daki eksik değerleri "No" ile dolduracağız. Veri seti belgelerine göre, Bsmt Exposure'için "Hayır", "Maruziyet Yok" anlamına gelmektedir.

<br/>

Şimdi BsmtFin Tip 1, Bsmt Qual ve Bsmt Cond null değilken BsmtFin Tip 2'nin null olduğu satıra bir göz atalım:

```python
dataset[~pd.isna(dataset['Bsmt Cond']) &
        pd.isna(dataset['BsmtFin Type 2'])][bsmt_columns]
```

Bu satır için BsmtFin Tip 2'deki eksik değeri "Unf" ile dolduracağız. Veri seti belgelerine göre, BsmtFin Type 2 için "Unf" "Bitmemiş" anlamına gelir:

```python
for col in ["Bsmt Half Bath", "Bsmt Full Bath", "Total Bsmt SF",
            "Bsmt Unf SF", "BsmtFin SF 2", "BsmtFin SF 1"]:
    dataset[col].fillna(0, inplace=True)

dataset.loc[~pd.isna(dataset['Bsmt Cond']) &
            pd.isna(dataset['Bsmt Exposure']), "Bsmt Exposure"] = "No"
dataset.loc[~pd.isna(dataset['Bsmt Cond']) &
            pd.isna(dataset['BsmtFin Type 2']), "BsmtFin Type 2"] = "Unf"

for col in ["Bsmt Exposure", "BsmtFin Type 2",
            "BsmtFin Type 1", "Bsmt Qual", "Bsmt Cond"]:
    dataset[col].fillna("No Basement", inplace=True)
```

**Mas Vnr Area ve Mas Vnr Type**

---

Bu iki sütunun her birinde 23 eksik değer bulunmaktadır. Bu eksik değerleri Mas Vnr Tipi için "Yok" ve Mas Vnr Alanı için 0 ile dolduracağız. Mas Vnr Tipi için "Yok" ifadesini kullanacağız çünkü veri seti dokümantasyonunda Mas Vnr Tipi için "Yok" ifadesi "Yok" (yani duvar kaplaması yok) anlamına gelmektedir:


```python
dataset['Mas Vnr Area'].fillna(0, inplace=True)
dataset['Mas Vnr Type'].fillna("None", inplace=True)
```

Bu fonksiyon, ortalamayı çıkararak ve birim varyansa ölçeklendirerek özellikleri standartlaştırır. Her özellik üzerinde bağımsız olarak çalışır. Bir x değeri için

X_train ve X_test özellikleri (tahmin edicileri); y_train ve y_test ise hedefi temsil etmektedir. Scikit-Learn kullanarak, örneğin aşağıdaki gibi bir karar ağacı modeli oluşturabiliriz:

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=14, min_samples_split=5, max_features=20)
```

Bunu yapabiliriz ancak max_depth, min_samples_split ve max_features parametreleri için daha iyi değerler seçersek muhtemelen daha iyi bir performans elde ederiz. Bunu yapmak için birçok parametre kombinasyonunu inceleyeceğiz ve en iyi puanı veren kombinasyonu seçeceğiz. Scikit-Learn bu amaç için kullanışlı bir fonksiyon sunuyor: GridSearchCV(). Yukarıdaki örnek için aşağıdakileri yapacağız:

Bazen, parametre kombinasyonlarının sayısı fazla olduğunda, GridSearchCV()'nin çalışması çok uzun sürebilir. Bu nedenle GridSearchCV()'ye ek olarak, bazen GridSearchCV()'ye benzer olan RandomizedSearchCV()'yi kullanırız, ancak tüm parametre kombinasyonlarını kullanmak yerine n_iter tarafından belirtilen sayıda rastgele kombinasyon seçer. Yukarıdaki örnek için RandomizedSearchCV() işlevini aşağıdaki gibi kullanabiliriz:

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor

clf = RandomizedSearchCV(DecisionTreeRegressor(), parameter_space, cv=4,
                         scoring="neg_mean_absolute_error", n_iter=100)
```

Bu, RandomizedSearchCV() işlevinin 100 parametre kombinasyonunu rastgele seçmesini sağlayacaktır.

Linear Regression :

Doğrusal Regresyon için üç algoritmik uygulama seçeceğiz: Ridge Regression ve Elastic Net. Bu algoritmaların Scikit-Learn paketinde sağlanan uygulamalarını kullanacağız.

Bu model aşağıdaki sözdizimine sahiptir:


```python
 from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0, fit_intercept=True, solver='auto', random_state=None)

```

İlk olarak, bizim tarafımızdan sağlanan bir parametre uzayında en iyi model parametrelerini aramak için GridSearchCV() kullanacağız. alpha parametresi düzenlileştirme gücünü temsil eder, fit_intercept bu model için kesişimin hesaplanıp hesaplanmayacağını belirler ve solver hesaplama rutinlerinde hangi çözücünün kullanılacağını kontrol eder.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

parameter_space = {
    "alpha": [1, 10, 100, 290, 500],
    "fit_intercept": [True, False],
    "solver": ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
}

clf = GridSearchCV(Ridge(random_state=3), parameter_space, n_jobs=4,
                   cv=3, scoring="neg_mean_absolute_error")

clf.fit(X_train, y_train)
print("Best parameters:")
print(clf.best_params_)
```

Seçilen parametreler için makul değerler kullanarak yukarıdaki parametre uzayını tanımladık. Daha sonra 3 katlama ile GridSearchCV() kullandık (cv=3). Şimdi Ridge modelimizi bulunan en iyi parametrelerle oluşturuyoruz:

```python
ridge_model = Ridge(random_state=3, **clf.best_params_)
```

Ardından eğitim setimizi (X_train ve y_train) kullanarak modelimizi eğitiriz:

```python
ridge_model.fit(X_train, y_train);
```

Son olarak, modelimizi X_test üzerinde test ediyoruz. Ardından, yukarıda açıkladığımız gibi MAE metriğini kullanarak tahminlerini y_test'teki gerçek değerlerle karşılaştırarak model performansını değerlendiriyoruz:

```python
from sklearn.metrics import mean_absolute_error

y_pred = ridge_model.predict(X_test)
ridge_mae = mean_absolute_error(y_test, y_pred)
print("Ridge MAE =", ridge_mae)
```

# **Ortak Önemli Özellikler**

---
Şimdi, hem XGBoost hem de Grandient Boosting modelleri için hangi özelliklerin en önemli özellikler arasında olduğunu görelim ve iki model açısından önemlerindeki farkı bulalım:


```python
common_imp_feat = [x for x in xgb_feature_importances.index
                   if x in gb_feature_importances.index]
commImpFeat_xgb_scores = [xgb_feature_importances[x]
                          for x in common_imp_feat]
commImpFeat_gb_scores = [gb_feature_importances[x]
                         for x in common_imp_feat]

ind = np.arange(len(commImpFeat_xgb_scores))
width = 0.35

fig, ax = plt.subplots()
ax.bar(ind - width/2, commImpFeat_xgb_scores, width,
       color='#003f5c', label='XGBoost');
ax.bar(ind + width/2, commImpFeat_gb_scores, width,
       color='#ffa600', label='Gradient Boosting')
ax.set_xticks(ind);
ax.set_xticklabels(common_imp_feat);
ax.legend();
plt.xticks(rotation=90);
```
